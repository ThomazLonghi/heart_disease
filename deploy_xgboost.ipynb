# **DESAFIO 1**

Este algoritmo realiza a previsão de uma pessoa ter tendência a desenvolver algum tipo de doença cardíaca com base em alguns fatores clínicos e laboratoriais

## **Abertura dos atribuos previsores e alvo**

```python
import pandas as pd
import numpy as np
```

```python
previsores = pd.read_csv('/content/drive/MyDrive/curso/previsores', header=None)
```

```python
previsores
```

```python
alvo = pd.read_csv('/content/drive/MyDrive/curso/alvo', header=None)
```

```python
alvo
```

## **Carregando Classficador XGBoost**

```python
from xgboost import XGBClassifier
xg = XGBClassifier(max_depth=2, learning_rate=0.02, n_estimators=250, objective='binary:logistic', random_state=3)
xg.fit(previsores, alvo)
```

## **Simulação de novos pacientes**

```python
paciente1 = [40, 0, 1, 140, 289, 0, 0, 172, 0, 0, 0]
```

```python
paciente1 = np.asarray(paciente1).reshape(1,-1)
```

```python
resultado = xg.predict(paciente1)

if resultado == 1:
  print('Paciente POSSUI tendência de problemas cardíacos')
else:
  print('Paciente NÃO POSSUI tendência de problemas cadíacos')
```
